```python
import pandas as pd
import os
import shutil
import random
```

```python
# Create the data for positive samples
FILE_PATH = "chestxray/metadata.csv"
IMAGES_PATH = "chestxray/images"
```

```python
df = pd.read_csv(FILE_PATH)
df.shape
```

```python
df.head()
```

```python
# Create COVID Dataset
TARGET_DIR = "Dataset/Covid"
if not os.path.exists(TARGET_DIR):
    os.mkdir(TARGET_DIR)
    print("Covid folder created")
```

```python
# Check the number of covid xrays and copy them to 'Dataset/Covid'
covid_images = 0
for (i, row) in df.iterrows():
    # Check if the images is covid positive and front view
    if (row["finding"] == "Pneumonia/Viral/COVID-19" and row['view'] == "PA"):
        # Select the file and copy it to new location
        filename = row['filename']
        image_path = os.path.join(IMAGES_PATH, filename)
        image_copy_path = os.path.join(TARGET_DIR, filename)
        shutil.copy2(image_path, image_copy_path)
        covid_images += 1
print("Normal Images Copied...")
```

```python
# Sampling of Images from Kaggle
KAGGLE_FILE_PATH = "chest_xray_kaggle/train/NORMAL"
TARGET_NORMAL_DIR = "Dataset/Normal"
```

We have to select normal images that are equivelant of COVID images.
For this 
- We will take the names of all the Normal Images.
- Shuffle these images names
- Select the first n images and copy them to target folder


```python
# Get the list of Images names
image_names = os.listdir(KAGGLE_FILE_PATH)

# Shuffle the names
random.shuffle(image_names)

# Select and copy the first n images
for i in range(covid_images):
    normal_filename = image_names[i]
    image_path = os.path.join(KAGGLE_FILE_PATH, normal_filename)
    target_path = os.path.join(TARGET_NORMAL_DIR, normal_filename)
    shutil.copy2(image_path, target_path)
print("Normal Images Copied...")
```
